```python
import numpy as np
import pandas as pd

cities = [f'City_{i+1}' for i in range(10)]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

np.random.seed(42)
temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))

df = pd.DataFrame(data=temperature_data, index=cities, columns=months)

print(df.head())
```

```python
df['Annual_Avg_Temperature'] = df.mean(axis=1)

highest_avg_city = df['Annual_Avg_Temperature'].idxmax()
lowest_avg_city = df['Annual_Avg_Temperature'].idxmin()

print(f'City with the highest average temperature: {highest_avg_city}')
print(f'City with the lowest average temperature: {lowest_avg_city}')
print(df[['Annual_Avg_Temperature']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
for city in cities:
    plt.plot(df.columns[:-1], df.loc[city, df.columns[:-1]], marker='o', label=city)

plt.title('Monthly Temperature Trends for Each City')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.legend(title='Cities', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Annual_Avg_Temperature'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Annual Average Temperatures')
plt.xlabel('Annual Average Temperature (Â°C)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
