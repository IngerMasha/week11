```python
import numpy as np

matrix = np.random.rand(4, 4)
print("Original Matrix:\n",  np.round(matrix, 2))

diagonal_elements = np.diag(matrix)
print("Diagonal Elements:\n",  np.round(diagonal_elements, 2))

diagonal_matrix = np.diag(diagonal_elements)
print("Diagonal Matrix:\n",  np.round(diagonal_elements, 2))
```

```python
array = np.random.rand(20)
print("Original Array:\n", np.round(array, 2))

array[array > 0.5] = -1
print("Modified Array:\n", np.round(array, 2))
```

```python
array = np.random.rand(10)
print("Original Array:\n", array)

mean = np.mean(array)
std_dev = np.std(array)

normalized_array = (array - mean) / std_dev
print("Normalized Array:\n", normalized_array)
```

```python
array1 = np.random.rand(10)
array2 = np.random.rand(10)

print("Array 1:\n", array1)
print("Array 2:\n", array2)

correlation_matrix = np.corrcoef(array1, array2)
pearson_correlation = correlation_matrix[0, 1]

print("Pearson Correlation Coefficient:\n", pearson_correlation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)  # For reproducibility
days = np.arange(15)
values = np.random.rand(15)

mean_days = np.mean(days)
mean_values = np.mean(values)

numerator = np.sum((days - mean_days) * (values - mean_values))
denominator = np.sqrt(np.sum((days - mean_days)**2) * np.sum((values - mean_values)**2))

m = numerator / denominator
b = mean_values - m * mean_days

trend_line = m * days + b

plt.figure(figsize=(10, 6))
plt.plot(days, values, 'o', label='Time Series Data')
plt.plot(days, trend_line, '-', label='Trend Line', color='red')
plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Time Series Data and Trend Line')
plt.legend()
plt.grid(True)
plt.show()
```
