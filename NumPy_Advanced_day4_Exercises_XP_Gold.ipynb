```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from sklearn.linear_model import LinearRegression

matrix1 = np.random.rand(5, 5)
print("Matrix:\n", matrix1)

eigenvalues, eigenvectors = np.linalg.eig(matrix1)
print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

matrix1_mean = np.mean(matrix1)
matrix1_std = np.std(matrix1)
normalized_matrix1 = (matrix1 - matrix1_mean) / matrix1_std
print("Normalized Matrix:\n", normalized_matrix1)
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

data_skewness = skew(data)
data_kurtosis = kurtosis(data)
print(f"Skewness: {data_skewness}")
print(f"Kurtosis: {data_kurtosis}")
```

```python

monthly_sales_data = np.random.randint(100, 500, size=12)
print("Monthly Sales Data:", monthly_sales_data)

months = np.arange(1, 13).reshape(-1, 1)
model = LinearRegression().fit(months, monthly_sales_data)
future_months = np.arange(13, 25).reshape(-1, 1)
forecasted_sales = model.predict(future_months)
print("Forecasted Sales for Next 12 Months:", forecasted_sales)

plt.plot(months, monthly_sales_data, label='Actual Sales')
plt.plot(future_months, forecasted_sales, label='Forecasted Sales', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data and Forecast')
plt.legend()
plt.show()
```

```python
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B'],
    'Sales': [200, 150, 300, 250, 180],
    'Month': ['January', 'February', 'March', 'April', 'May']
}
df = pd.DataFrame(data)

grouped_df = df.groupby('Product').agg({'Sales': ['sum', 'mean', 'max']})
print("Grouped and Aggregated Data:\n", grouped_df)
```
